```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
column_names = ["EP","ACC","SE","SP","PC","F1","JS","DC"]
rt = pd.read_csv("Rtraining_plot.csv",header=None,names=column_names)
rv = pd.read_csv("Rvalidation_plot.csv",header=None,names=column_names)
ut = pd.read_csv("Utraining_plot.csv",header=None,names=column_names)
uv = pd.read_csv("Uvalidation_plot.csv",header=None,names=column_names)
```

```python
x = ut["EP"]
y1 = ut["DC"]
y2 = rt["DC"]
plt.plot(x,y1,'red',label="U_Net")
plt.plot(x,y2,'blue',label="R2U_Net")
plt.legend(loc='upper right')
plt.ylim(0.5,1)
plt.grid(color='gray', linestyle='-.', linewidth=0.5)
plt.xlabel('Epochs')
plt.title('Training accuracy')
plt.savefig("trainacc025.png")
plt.show()
```

```python
x = uv["EP"]
y1 = uv["DC"]
y2 = rv["DC"]
plt.plot(x,y1,'red',label="U_Net")
plt.plot(x,y2,'blue',label="R2U_Net")
plt.legend(loc='upper right')
plt.ylim(0,1)
plt.grid(color='gray', linestyle='-.', linewidth=0.5)
plt.xlabel('Epochs')
plt.title('Validation accuracy')
plt.savefig("validacc025.png")
plt.show()
```
